6) Please write a program which asks the user for a number of days. The program 
then prints out the number of seconds in the number of days given. 
How many days? 1

7) take number of persons 
then number of persons in each group 
ex: persons=11
groups=3
then no of each group = 4

persons = 8
groups = 4
then no in each is = 2

```python
persons = input("Enter total number of persons: ")
per_group = input("Enter number of persons per group: ")
print(f"Total persons: {persons}, Persons per group: {per_group}")

in_each_group = int(persons) // int(per_group)
in_each_group += ((int(persons) % int(per_group) ) >= 1) 
print(f"Each group will have {in_each_group} persons")
```
