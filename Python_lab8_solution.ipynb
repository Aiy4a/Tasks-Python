TASK manager system 


**With Bonus**

```python
class Task:
    def __init__(self, name):
        self.name = name
        self.status = False
        self.priority = 5       #priority ranges between 1 and 10


class TaskManager:
    def __init__(self):
        self.tasks_list = []


    #1. Create Tasks
    def create_task(self, task_name):
        """Create the task and set status incomplete, priority = 5 as default"""
         #Bonus
        for task in self.tasks_list:
            if task.name == task_name:
                print(f"Invalid! {task_name} is already in list, Use another name")
                return
        task = Task(task_name)
        self.tasks_list.append(task)


    #2. Mark a Task as Completed 
    def task_complete(self, task_name, status_to_update = True):
        """Modify the task status to complete"""
        for task in self.tasks_list:
            if task.name == task_name:
                if status_to_update not in [True, False]:
                    print("Invalid status. Please use True or False.")
                    return
                elif task.status == True:
                    return
                else:
                    task.status = status_to_update
                    return
        print("Task not found.")


    #3. List All Tasks 
    def display_tasks_info(self):
        """Prints all tasks with their informations"""
        for task in self.tasks_list:
            print(f"Task: {task.name}, Priority is {task.priority} , Completed: {'Yes' if task.status else 'No'}")


    #4. List Pending Tasks
    def pending_tasks(self):
        """Lists all the incomplete tasks"""
        count = 0
        print("Pending Tasks: ")
        for task in self.tasks_list:
            if task.status == False:
                count += 1
                print(f"{count}) {task.name} ")


    #Bonus
    def count_complete(self):
        """Count completed vs pending tasks"""
        count = 0
        for task in self.tasks_list:
            if task.status == False:
                count += 1
        print(f"Number of pending tasks: {count} ")
        print(f"Number of completed tasks: {len(self.tasks_list)-count} ")


    #managing priority
    def add_priority(self , task_name , priority):
        """Add priority to task from 1 --> 10 
        where 1 for least priority 
        and  10 for the highest priority
        """
        if priority < 0 and priority > 10:
            print("Invalid! priority anges between 1 and 10 ")
        else:
            for task in self.tasks_list:
                if task.name == task_name:
                    task.priority = priority
    

    def sort_by_priority(self):
        """Sort the tasks from highest to lowest priority"""
        self.tasks_list.sort(key=lambda task: task.priority , reverse = True) 
        #Call sort(), pass a key function that tells sort() how to compare items.
        print("Tasks sorted by priority:")
        for task in self.tasks_list:
            if task.status == False: # To print only pending tasks
                print(f"Task: {task.name}, Priority: {task.priority}")



tasks= TaskManager()
tasks.create_task("Do laundry")
tasks.create_task("Do laundry")
tasks.create_task("Make a cake")
tasks.create_task("Do the dishes")
tasks.create_task("Do Homework")

print(f"the list now contains these tasks {[task.name for task in tasks.tasks_list]}")

tasks.display_tasks_info()
print("\n")

tasks.task_complete("Do laundry", True)
tasks.display_tasks_info()
print("\n")

tasks.pending_tasks()
print("\n")

tasks.count_complete()
print("\n")
tasks.add_priority("Make a cake" , 3)
tasks.add_priority("Do the dishes" , 8)
tasks.add_priority("Do Homework" , 4)
tasks.display_tasks_info()
print("\n")
tasks.sort_by_priority()
```
