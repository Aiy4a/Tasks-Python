1) write a program which asks the user to choose between addition and removal. 
Depending on the choice, the program adds an item to or removes an item from the end of 
a list. The item that is added must always be one greater than the last item in the list. The 
first item to be added must be 1.

```python
numbers = []
count = 0
while True:
    edit = input("Enter a(d)d, (r)emove or e(x)it: ")

    if edit == "r":
        if len(numbers) == 0:
            print("Invalid input try again")
        else:
            del numbers[-1]
            count -= 1
            print(f"Your list now is: {numbers}")
    elif edit == "d":
        count += 1
        numbers.append(count)
        print(f"Your list now is: {numbers}")
    elif edit == "x":
        print("\nBye!")
        break 
    else:
         print("Invalid input try again")


```

2) write a function named length_of_longest, which takes a list of strings as its 
argument. The function returns the length of the longest string.

```python
def length_of_longest(my_list):
    """Get the lenght of the longest string in a list"""
    lengths = [len(word) for word in my_list]
    return max(lengths)



my_list1 = ["first", "second", "fourth", "eleventh"] 
result = length_of_longest(my_list1) 
print(f"Length of the longest is: {result}")

my_list2 = ["adele", "mark", "dorothy", "tim", "hedy", "richard"] 
result = length_of_longest(my_list2) 
print(f"Length of the longest is: {result}")
```

3) write a function named all_the_longest, which takes a list of strings as its argument. 
The function should return a new list containing the longest string in the original list. If 
more than one are equally long, the function should return all of the longest strings. 
The order of the strings in the returned list should be the same as in the original. 

```python
def all_the_longest(my_list):
    """Get all the longest strings in a list"""
    maximum = length_of_longest(my_list)
    return [strings for strings in my_list if len(strings) == maximum ]
    
    

my_list1 = ["first", "second", "fourth", "eleventh"] 
result = all_the_longest(my_list1) 
print(f"Longest strings are: {result}")

my_list2 = ["adele", "mark", "dorothy", "tim", "hedy", "richard"] 
result = all_the_longest(my_list2) 
print(f"Longest strings are: {result}")
```

4) write a program which asks the user for words. If the user types in a word for the 
second time, the program should print out the number of different words typed in, and exit. 

```python
words = []
count = 0 
while True:
    word = input("Please enter your word or 'q' to Quit: ")
    if word in words or word == "q": 
        print(f"\nYou typed in {count} different words ")
        break
    else:
        count += 1 
        words.append(word)
        print(f"Word: {word}")
    
```
