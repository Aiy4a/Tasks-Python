```python
class Restaurant:
    """A class to represent restaurant name and cuisine"""
    def __init__(self, name, cuisine_type):
        """Initialize name and cuisine type attributes."""
        self.name = name
        self.cuisine_type = cuisine_type
        self.number_served = 0

    def describe_restaurant(self):
        print(f"Restaurant Name: {self.name} and the cuisine type is: {self.cuisine_type}")

    def open_restaurant(self):
        print(f"{self.name} is open!")

    def set_number_served(self, number):
        if number >= self.number_served:
            self.number_served = number
        else:
            print("You can't decrease the number of served customers.")

    def increment_number_served(self, number):
        if number > 0:
            self.number_served += number
        else:
            print("You can only increment by a positive number.")



restaurant1 = Restaurant("Pastawisi", "Italian")
restaurant2 = Restaurant("Sushi", "Japanese")

print(f"{restaurant1.number_served}")
restaurant1.set_number_served(10)
print(f"{restaurant1.number_served}")

restaurant1.increment_number_served(5)
print(f"{restaurant1.number_served}")

restaurant1.increment_number_served(-3)
print(f"{restaurant1.number_served}")
```
