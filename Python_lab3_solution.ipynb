1) Please write a function named triangle, which draws a triangle of 
hashes, and takes one argument. The triangle should be as tall and 
as wide as the value of the argument.

```python
def triangle(number):
    """Draws a triangle of hashes"""
    for i in range(number + 1):
        print(i * '#')
    

triangle(5)
```

2) Please write a function named greatest_number, which takes three 
arguments. The function returns the greatest in value of the three. 
[without using max function, and solve it in two ways]

**Way 1**

```python
def greatest_no(num1 , num2 , num3):
    """Return the greatest value out of three numbers"""
    if num1 >= num2 and num1 >= num3:
        return num1
    if num2 >= num1  and num2 >= num3:
        return num2
    if num3 >= num1  and num3 >= num2:
        return num3
    

max1 = greatest_no(0,0,0)
print(f"{max1}")
```

max(a,b)=  [(a+b) / 2] + [(|a-b|) / 2]

max(a,b,c)= [(max(a,b) + c) / 2] + [(|max(a,b) - c|) / 2] 


```python
def max_of_two(a, b):
    """Return the greatest value out of two numbers"""
    formula_ = (a + b) / 2 + abs(a - b) / 2
    return formula_

def max_of_three(a, b, c):
    """Return the greatest value out of three numbers"""
    formula_ = max_of_two(max_of_two(a, b), c)
    return formula_

max_no = int(max_of_three(0, 10, 20))
print(f"{max_no}")
```

3) Write a function sum_series(n) that calculates the sum of the series: 

```python
def sum_series(n):
    """Calculates the sum of a series of numbers"""
    sum = 0
    for i in range( 1 , n+1  ):
        if i % 2 != 0:   #odd
            sum += i
        else:
            sum -= i
    print(f"{sum}")

sum_series(3)
```

4) Write a Python function to calculate the factorial of a number (a 
non-negative integer). The function accepts the number as an 
argument.

way1

```python
def factorial_no1(num):
    """Calculates the factorial of a number"""
    if num < 0:
        print("Invalid")
    else:
        result = 1
        for i in range(1, num + 1):
                result *= i
        return result

factorial_no1(7)
```

way2  (recursive)

```python
def factorial_no2(num):
    """Calculates the factorial of a number"""
    if num < 0:
            print("Invalid")
        
    elif num == 0 or num == 1:
        return 1
    else:
        return num * factorial_no2(num - 1)


factorial_no2(7)
```

5) Implement a function that checks if a given word is a palindrome 
(reads the same forwards and backwards)

```python
def check_palindrome(word):
    """Checks if a word is a palindrome"""
    word = word.lower()
    letters = len(word)

    for i in range(letters // 2):
        if word[i] != word[letters - (i+1)]: 
            print("Not a palindrome")
            return
        
    print("It's a palindrome")
            


check_palindrome("Madam")
```
