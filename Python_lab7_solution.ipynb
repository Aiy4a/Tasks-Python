Write a Python program that: 
1. Asks the user for a file name. 
2. Reads the contents of the file. 
3. Handles the following exception:  
o FileNotFoundError if the file does not exist. 
4. If the file is read successfully, print its contents. 
5. Include an option for the user to try again if an error occurs. 

```python
from pathlib import Path

def file_reader():
    """Takes a file path as input and prints its contents if found"""
    path = input("Enter the file path: ")
    path = Path(path)
    try:
        content = path.read_text()
    except FileNotFoundError:
        print(f"Error: File '{path}' not found.")
        user_input = input("Would you like to try again? (y/n)?").lower()
        if user_input == 'y':
            file_reader()
        else:
            return
    else:
        print(f"Contents of '{path}': ")   
        print(content) 

file_reader()
```

Write a Python program that: 
1. Prompts the user to enter their name, age, and email. 

2. Validates the inputs as follows: 

    o Name must only contain alphabetic characters and spaces. 

    o Age must be a positive integer.
     
    o Email must follow a standard email format (e.g., contain "@" and "."). 

3. Handles the following exceptions:  
    o  ValueError if the age is not an integer. 

4. Outputs the validated data.

```python
################ to check email validity ##################
def email_check(email):
    """Validate the email format entered by the user"""
    while(
        "@" not in email or 
        "." not in email.split("@")[-1] or
        email.count("@") != 1 or 
        not any(ch.isalpha() for ch in email.split("@")[-1]) or ## this two rows to check if there is alpha before
        not any(ch.isalnum() for ch in email.split("@")[0]) or     ## and after @ "handle cases like @.com or .@"
        email.split("@")[0].startswith(".") or
        email.split("@")[-1].endswith(".") or
        ".." in email or
        "@." in email or
        ".@" in email

    ):
        print("ERROR: Enter a valid email address.")
        email = input("Enter your email or 'q' to quit: ")
        if email.lower() == 'q':
            return None 

################ to check name validity ####################
def user_name_check(name):
    """Validate the user name format entered by the user"""
    while not all(part.isalpha() for part in name.split()):
        print("ERROR: Enter alphabetic characters only.")
        name = input("Enter your name or 'q' to quit: ")
        if name.lower() == 'q':
            return None
    return name

################ to check age validity ####################
def age_check(age):
    """Validate the user age format entered by the user"""
    try:
        age = int(age)
    except ValueError:
        print("ERROR:Age must be an integer(ValueError)!")
        age = input("Enter your age: ")
    else:
        while  not str(age).isdigit() or age <= 0:
            print("ERROR:Age must be a valid positive integer!")
            age = input("Enter your age or 'q' to quit: ")
            if age.lower() == 'q':
                return None
    return age


######### To validate all user inputs #########
def validate_user_input():
    """Validates user inputs for name, age, and email"""
    # Validate name (alphabetic characters only)
    name = input("Enter your name or 'q' to quit: ")
    if name.lower() != 'q':
       name = user_name_check(name)
    else:
        return
    

    # Validate age (positive integer)
    age = input("Enter your age or 'q' to quit: ")
    if age.lower() != 'q':
        age = age_check(age)
    else:
        return
    
    # Validate email by calling email_check function
    email = input("Enter your email or 'q' to quit: ")
    if email.lower() != 'q':
        email = email_check(email)  
    else:
        return
    
    print(f"Name: {name}\nAge: {age}\nEmail: {email}")


validate_user_input()
```

Enhance Problem 3 by allowing the user to input their own data (e.g., name, age, and city) 
instead of using hardcoded values. Include validation to ensure: 
1. Name is a string. 
2. Age is a positive integer. 
3. City is a non-empty string. 


```python
import json 

def data_store_bonus( file_name = "data.json" , data={}):
    """Saves user data into a file in jason format"""
    path = Path(file_name + ".json")
    if data == {}:
        name = input("Enter name or 'q' to quit: ")
        if name.lower() == 'q':
            return None
        name = user_name_check(name)
        data['name'] = name

        age = input("Enter age or 'q' to quit: ")
        if age.lower() == 'q':
            return None
        age = age_check(age)    
        data['age'] = age

        city = input("Enter city: ") 
        while city == "":
            print("ERROR: City cannot be empty.")
            city = input("Enter city or 'q' to quit: ")
            if city.lower() == 'q':
                return None

        data['city'] = city

    try:
        json.dump(data, path.open('w')) #this opens file and takes data to write in it 
    except IOError:
        print(f"Error: {file_name} could not be written.")
    else:
        print(f"Data saved successfully to '{file_name}'")
        print(f"Contents of '{file_name}': ")
        print(json.load(path.open('r')))  # this opens file and reads data from it 
        


data_store_bonus()
```
