```python
def readposint():
    while True:
        number = input("Enter a positive integer or 'q' to quit: ")
        if number == "q":
            break
        try:
            number = int(number)
        except KeyboardInterrupt:
            print("Input was interrupted and canceled")
            return
        except ValueError:
            print("Invalid input. Enter a valid positive integer.")
            
        else:
            if number <= 0:
                print("The number is not positive. Try again.")
            else:
                print(f"You entered the positive integer: {number}")


readposint()
```
